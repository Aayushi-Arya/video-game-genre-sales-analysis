```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r"C:\Users\aayus\Documents\quantlake\vgsales.csv")

# Group sales by Genre
genre_sales = df.groupby("Genre")[["Global_Sales", "JP_Sales"]].sum().sort_values("Global_Sales", ascending=False)

# Create Global Sales by Genre chart
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_sales.index, y=genre_sales["Global_Sales"])
plt.title("Global Video Game Sales by Genre")
plt.xlabel("Genre")
plt.ylabel("Global Sales (millions)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("global_sales_by_genre.png")
plt.show()

plt.close()

# Create Japan Sales by Genre chart
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_sales.index, y=genre_sales["JP_Sales"])
plt.title("Japan Video Game Sales by Genre")
plt.xlabel("Genre")
plt.ylabel("Sales in Japan (millions)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("japan_sales_by_genre.png")
plt.show()

plt.close()
```
